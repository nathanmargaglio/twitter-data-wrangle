```python
import tweepy
import json
import pandas as pd
from time import time
import sys

# import Twitter archive data
twitter_archive_df = pd.read_csv('twitter-archive-enhanced.csv')

consumer_key = ''
consumer_secret = ''
access_token = ''
access_secret = ''

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth)
```

```python
# Get tweet data for each tweet in the archive
twitter_api_data = []

start_time = time()
prev_time = start_time

for twt in twitter_archive_df.iterrows():
    try:
        tweet = api.get_status(twt[1].tweet_id, 
                               tweet_mode='extended', 
                               wait_on_rate_limit=True, 
                               wait_on_rate_limit_notify=True)
        twitter_api_data.append(tweet._json)

        new_time = time()
        print('ID:{} ({:0.2f} sec, {:0.2f} sec total)'.format(
            twt[1].tweet_id,
            new_time - prev_time,
            new_time - start_time))
        prev_time = new_time
    except:
        e = sys.exc_info()[0]
        print(e)
        
with open('tweet_json.txt', 'w') as outfile:
    json.dump(twitter_api_data, outfile)
```
